## Huggingface API

### Gemma

```python
from google.colab import userdata
Hugginface_APIKEY= userdata.get('HUGGINGFACE_TOKEN')
```

```python
model_id='google/gemma-7b'
```

```python
import requests
```

```python
API_URL="https://api-inference.huggingface.co/models/google/gemma-7b"
```

```python
headers = {"Authorization":f"Bearer {Hugginface_APIKEY}"}
```

```python
def query(prompt):
  response = requests.post(API_URL, headers = headers, json = prompt)
  print(response)
  return response.json()
```

```python
question={ "inputs":"can you please let me know a capital of india" }
```

```python
output= query(question)
output[0]
```

```python
print(output[0]['generated_text'])
```

## Gemini API (Gemini)

```python
from google.colab import userdata
Gemini_API= userdata.get('GEMINI_API')
```

### Gemini-Pro (Text Generation Model)

```python
import google.generativeai as genai
```

```python
genai.configure(api_key=Gemini_API)
```

```python
for models in genai.list_models():
  print(models)
```

```python
# printing the GenAI - content generation models

for models in genai.list_models():
  if "generateContent" in models.supported_generation_methods:
    print(models.name)
```

```python
text_model= genai.GenerativeModel("gemini-1.0-pro")
```

```python
response= text_model.generate_content("what is a capital of Tamil Nadu and give me a sweet poem on Tamil Nadu")
```

```python
print(response.text)
```

```python
response.candidates
```

```python
response.parts[0]
```
